```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a:int, b:int):
    """
        It will return the sum of two numbers

        Args:
            a : Number One
            b : Number Two

    """
    return a + b

@tool
def multiply_numbers(a:int, b:int):
    """
        It will return the product of two numbers

        Args:
            a : Number One
            b : Number Two

    """
    return a * b
```

```python
add_numbers.invoke({"a":11,"b":9})
```

```python
from langchain_openai import ChatOpenAI
from langchain.agents import create_agent
```

```python
llm = ChatOpenAI(model="gpt-5-nano")
```

```python
agent = create_agent(
  model=llm,
  tools=[add_numbers,multiply_numbers],
  system_prompt="You are a math teacher, and always use tool for calculation."
)
```

```python
response = agent.invoke({"messages":[{"role":"user","content":"What is 2 + 3 * 3?"}]})
print(response["messages"][-1].content)
```
