```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI(model="gpt-5-nano", streaming=True)
```

```python
question = "What is Tuple?"
```

```python
response = llm.stream(question) # to get the content into chunks we use stream and invoke is used to get the content at one go/time # This line needs to be above for loop in the same cell to restore the memory. I you place it in the above cell it will erase the memory and you will not receive the output.
for chunk in response:
    print(chunk.content, end="")
```
